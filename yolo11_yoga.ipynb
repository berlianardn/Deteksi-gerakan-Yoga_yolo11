```python
from ultralytics import YOLO

# Load model YOLO11‑pose (varian Nano, Small, Medium, Large, atau X)
model = YOLO("yolo11n-pose.pt")  

# Training model
model.train(
    data="datasets/data.yaml",     
    epochs=64,
    imgsz=640,
    batch=16,
    name="yolo11_yoga_pose", 
    task="pose",
    device=0
)
```

```python
from ultralytics import YOLO
import cv2

# Load model YOLO11-pose hasil training deteksi yoga
model = YOLO("best.pt")  # Pastikan model ini adalah YOLO11 pose model (misal hasil training YOLO11)

# Load video yoga
video_path = "testing5.mp4"
cap = cv2.VideoCapture(video_path)

# Info video
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))

# Output video
out = cv2.VideoWriter("output_yoga_pose5.mp4", cv2.VideoWriter_fourcc(*'mp4v'), fps, (width, height))

while cap.isOpened():
    success, frame = cap.read()
    if not success:
        break

    # Prediksi keypoints pose dengan YOLO11
    results = model.predict(source=frame, save=False, conf=0.3)

    # Ambil hasil prediksi
    for r in results:
        # Pastikan keypoints ada dan tidak kosong
        if r.keypoints is not None and r.keypoints.xy is not None and r.keypoints.xy.shape[0] > 0:
            keypoints = r.keypoints.xy[0].cpu().numpy()  # Ambil keypoints objek pertama
            print("Keypoints:", keypoints)

            # TODO: Tambahkan logika deteksi pose yoga berdasarkan keypoints
        else:
            print("⚠️ Tidak ada orang terdeteksi di frame ini.")

    # Plot hasil pose pada frame
    annotated_frame = results[0].plot()

    # Simpan frame ke video output
    out.write(annotated_frame)

    # Tampilkan frame dengan pose
    cv2.imshow("Yoga Pose Estimation", annotated_frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Bersihkan resource
cap.release()
out.release()
cv2.destroyAllWindows()
```
